```python
import logging

import yaml

from scrapers import RunwayScraper

logging.basicConfig(level=logging.INFO)
```

```python
with open('./interests/runway.yaml', mode='r') as fp:
    inventory_ids = yaml.safe_load(fp)
```

```python
for inventory_id in inventory_ids:
    try:
        runway_scraper = RunwayScraper(inventory_id)
        inventory = runway_scraper.scrape()
    except Exception as error:
        logging.error(error)
        logging.info(f'Skipping for inventory ID: {inventory_id}.')
    else:
        display(inventory)
```
