```python
import logging

import yaml

from scrapers import RunwayScraper

logging.basicConfig(level=logging.INFO)
```

```python
with open('./configs/runway.yaml', mode='r') as fp:
    item_ids = yaml.safe_load(fp)
for item_id in item_ids:
    try:
        runway_scraper = RunwayScraper(item_id)
        inventory = runway_scraper.scrape()
    except Exception as error:
        logging.error(error)
        logging.info(f'Skipping for inventory ID: {item_id}.')
    else:
        display(inventory)
```
